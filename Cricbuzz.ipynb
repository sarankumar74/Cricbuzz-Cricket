# Match over

```python
import requests

url = "https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/41881/overs"

headers = {
	"x-rapidapi-key": "6b8ab44a0emsh6af0e5875b56994p163df1jsncf3c5f60df98",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

data = response.json()
print(json.dumps(data,indent=4))
```

```python
import pandas as pd

rows = []

# ---- batsman & bowler ----
for key in ["batsmanStriker", "batsmanNonStriker", "bowlerStriker", "bowlerNonStriker"]:
    t = data["miniscore"].get(key, {})
    rows.append({
        'section': key,
        'id': t.get('id'),
        'name': t.get('name'),
        'nickName': t.get('nickName'),
        'runs': t.get('runs'),
        'balls': t.get('balls'),
        'fours': t.get('fours'),
        'strkRate': t.get('strkRate'),
        'overs': t.get('overs'),
        'wickets': t.get('wickets'),
        'economy': t.get('economy')
    })

# ---- innings scores ----
for t in data["miniscore"].get("inningsScores", {}).get("inningsScore", []):
    rows.append({
        'section': 'inningsScore',
        'inningsId': t.get('inningsId'),
        'runs': t.get('runs'),
        'wickets': t.get('wickets'),
        'overs': t.get('overs'),
        'balls': t.get('balls')
    })

# ---- performance ----
for t in data["miniscore"].get("performance", []):
    rows.append({
        'section': 'performance',
        'label': t.get('label'),
        'runs': t.get('runs'),
        'overs': t.get('overs'),
        'wickets': t.get('wickets')
    })

# ---- power play ----
for t in data["miniscore"].get("pp", {}).get("powerPlay", []):
    rows.append({
        'section': 'powerPlay',
        'ppId': t.get('ppId'),
        'ppOversFrom': t.get('ppOversFrom'),
        'ppOversTo': t.get('ppOversTo')
    })

# ---- bat team score ----
bat_score = data["miniscore"].get("batTeamScore", {})
rows.append({
    'section': 'batTeamScore',
    'teamScore': bat_score.get('teamScore'),
    'wickets': bat_score.get('wickets'),
    'overs': bat_score.get('overs'),
    'balls': bat_score.get('balls')
})

# ---- overs separation ----
for t in data.get("overSepList", {}).get("overSep", []):
    rows.append({
        'section': 'overSep',
        'overNum': t.get('overNum'),
        'batScore': t.get('batScore'),
        'overRuns': t.get('overRuns'),
        'overWickets': t.get('overWickets')
    })

# ---- match headers ----
header = data.get("matchHeaders", {})
rows.append({
    'section': 'matchHeader',
    'matchId': header.get('matchId'),
    'status': header.get('status'),
    'matchFormat': header.get('matchFormat'),
    'tossResults': header.get('tossResults'),
    'seriesName': header.get('seriesName')
})

# ---- mom players ----
for t in header.get("momPlayers", {}).get("player", []):
    rows.append({
        'section': 'momPlayer',
        'id': t.get('id'),
        'name': t.get('name'),
        'teamName': t.get('teamName')
    })

# ---- final DataFrame ----
df_all = pd.DataFrame(rows)
df_all
```
